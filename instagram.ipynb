```
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import math
from google.colab import drive

drive.mount('/content/drive')
instagram = nx.read_edgelist("/content/drive/My Drive/UFRPE/Periodos/PLE2020.4/Modelo/Projeto/grafo_instagram.txt", create_using=nx.DiGraph)

print()
print("Bem-vindo(a) à rede Instagram")
print()

print("Número de nós: ", instagram.number_of_nodes())
print("Número de enlaces: ", instagram.number_of_edges())

# Sequência do grau de entrada
degs_in = instagram.in_degree() 
deg_in = [d for n,d in degs_in]
print("Sequência do grau de entrada: ", deg_in)

# Grau médio de entrada
lambd = np.average(deg_in)
print("Grau médio de entrada: " + str(lambd))

# Grau máximo de entrada
hubs = max(degs_in, key=lambda item: item[1])
print("Grau do Hub de entrada: ", hubs[1])

# Sequência do grau de saída
degs_out = instagram.out_degree() 
deg_out = [d for n,d in degs_out]
print("Sequência do grau de saída: ", deg_out)

# Grau médio de saída
lambd = np.average(deg_out)
print("Grau médio de saída: " + str(lambd))

# Grau máximo de saída
hubs = max(degs_out, key=lambda item: item[1])
print("Grau do Hub de saída: ", hubs[1])

# Tamanho médio dos caminhos
print("Tamanho médio dos caminhos: ", nx.average_shortest_path_length(instagram))

# Eficiência dos enlaces
m = instagram.number_of_edges()
E = 1 - nx.average_shortest_path_length(instagram)/m
print("Eficiência dos enlaces: ", E)

# Coeficiente de aglomeração
print("Coeficiente de aglomeração: ", nx.average_clustering(instagram))

# Componente conectada
print("Tamanho da maior componente conectada: ", len(max(nx.weakly_connected_components(instagram), key=len)))

# Densidade
print("Densidade: ", nx.density(instagram))

# Diâmetro da componente fortemente conectado
insta = max(nx.strongly_connected_components(instagram), key=len)
SG = instagram.subgraph(insta).copy()
print("Diâmetro da rede: ", nx.diameter(SG))

# Histograma de entrada
plt.figure(1)
plt.hist(deg_in, weights= np.ones_like(deg_in) / len(deg_in))
plt.xlabel('Grau de entrada')
plt.ylabel('Probabilidade')
plt.show()

# Histograma de saída
plt.figure(2)
plt.hist(deg_out, weights= np.ones_like(deg_out) / len(deg_out))
plt.xlabel('Grau de saída')
plt.ylabel('Probabilidade')
plt.show()


```
